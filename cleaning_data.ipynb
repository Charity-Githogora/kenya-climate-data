```python
import pandas as pd
df = pd.read_csv('HKRE0_Nairobi_-_Doonholm.csv')
```

```python
df.shape
```

```python
# View the first few rows
df.head()

# Check column names and data types
df.info()

# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Dropping the columns that are completely empty
df = df.drop(columns=['snow', 'wdir', 'wpgt', 'tsun'])
```

```python
df_cleaned = df.copy()

# First forward-fill (ffill), then backward-fill (bfill)
df_cleaned = df_cleaned.ffill().bfill()
```

```python
# Verify cleaning
df_cleaned.isnull().sum()
```
